```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results_no_theta.csv')
df.head()
```

### Recall

```python
# Group by 'degree' and calculate the mean for the recall columns
recall_by_degree = df.groupby('n')[['linear_recall', 'random_recall', 'knn_recall']].mean()
colors = ['red', 'blue', 'green']
# Plot the recall values with dots for points
recall_by_degree.plot(kind='line', marker='o', figsize=(10, 6), color=colors)
plt.title('Recall of Methods vs. Size (n)')
plt.xlabel('n')
plt.ylabel('Recall')
plt.xscale('log', base=2)
plt.legend(title='Method')
plt.grid(True)
plt.show()
```

```python
# Group by 'degree' and calculate the mean for the recall columns
recall_by_degree = df.groupby('degree')[['linear_recall', 'random_recall', 'knn_recall']].mean()
colors = ['red', 'blue', 'green']
# Plot the recall values with dots for points
recall_by_degree.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Recall of Methods vs. Degree')
plt.xlabel('Degree')
plt.ylabel('Recall')
plt.xscale('log', base=2)
plt.legend(title='Method')
plt.grid(True)
plt.show()
```

```python
# Group by 'degree' and calculate the mean for the recall columns
recall_by_degree = df.groupby('dim')[['linear_recall', 'random_recall', 'knn_recall']].mean()
colors = ['red', 'blue', 'green']

# Plot the recall values with dots for points
recall_by_degree.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Recall of Methods vs. Dimension')
plt.xlabel('dimension')
plt.ylabel('Recall')
plt.xscale('log', base=2)
plt.legend(title='Method')
plt.grid(True)
plt.show()
```

```python
# Calculate the mean of the recall columns across the whole dataset
average_recall = df[['linear_recall', 'random_recall', 'knn_recall']].mean()

# Define colors for each bar
recall_colors = ['blue', 'orange', 'red']

# Plot the average recall with specific colors for each bar
ax = average_recall.plot(kind='bar', figsize=(8, 5), color=recall_colors)
plt.title('Overall Average Recall')
plt.xlabel('Method')
plt.ylabel('Average Recall')

# Add numbers on top of the bars
for i, value in enumerate(average_recall):
    ax.text(i, value + 0.01, f'{value:.2f}', ha='center', va='bottom')

plt.show()
```

```python
# Calculate the mean recall values grouped by 'epsnet_size_ratio'
recall_by_epsnet = df.groupby('epsnet_size_ratio')[['linear_recall', 'random_recall', 'knn_recall']].mean()

# Plot the recall values as a bar chart
ax = recall_by_epsnet.plot(kind='bar', figsize=(10, 6), color=colors)
ax.set_title('Recall of Methods vs. EpsNet Size Ratio')
ax.set_xlabel('EpsNet Size Ratio')
ax.set_ylabel('Recall')
ax.set_xticklabels(['1/128', '1/16'], rotation=0)  # Set custom x-axis labels
ax.legend(title='Method')
ax.grid(axis='y')
plt.show()
```
